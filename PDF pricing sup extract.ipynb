```python
# Creating a pdf file object
#pdfFileObj = open('BAYA82 PricingSupplement3009.pdf','rb')
#pdfFileObj = open('CLN862 PricingSupplement2007.pdf','rb')
pdfFileObj = open('IVC188-ApplicablePricingSupplement-02-12-2020.pdf','rb')
#pdfFileObj = open('Applicable Pricing Supplement - Class A2.pdf','rb')
#pdfFileObj = open('ASN885 PricingSupplement1207.pdf','rb')
#pdfFileObj = open('Applicable-Pricing-Supplement-SLI7-Execution.pdf','rb')
#pdfFileObj = open('snt06-applicable-pricing-supplement-11052022-executed.pdf','rb')
#pdfFileObj = open('NNF074 - PricingSupplement2611.pdf','rb')
#pdfFileObj = open('NNF082 PricingSupplement0806.pdf','rb')

```

### Creating a pdf reader object 

##### showing number of pahes


```python

pdfReader = PyPDF2.PdfFileReader(pdfFileObj)
# Getting number of pages in pdf file
pages = pdfReader.numPages
# Loop for reading all the Pages
pages
```

### Extracting, spitting and storing pages in a list

```python
list1=[]

for i in range(pages):
        # Creat2ing a page object
        pageObj = pdfReader.getPage(i)
        # Printing Page Number
        # Extracting text from page
        # And splitting it into chunks of lines
        text = pageObj.extractText().split("                   ")
        # Finally the lines are stored into list
        # For iterating over list a loop is used

        for i in range(len(text)):
                # Printing the line
                # Lines are seprated using "\n"
                X=text[i]
                list1.append(X)

        # For Seprating the Pages
# closing the pdf file object
pdfFileObj.close()
```

#### Splitting pages line by line and storing them in a list

```python
container=[]
for i in list1:
    
    i=i.split('\n')
    container.append(i)
```
